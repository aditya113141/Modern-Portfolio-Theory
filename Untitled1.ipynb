<h1>Modern Portfolio Theory</h1>

- Modern Portfolio Theory (MPT) is a Nobel Prize-winning economic theory. 
- It explains how risk-averse investors can construct portfolios to optimize or maximize expected return based on a given level of market risk. 
- <a href = "https://en.wikipedia.org/wiki/Harry_Markowitz">Harry Markowitz</a> pioneered this theory in his paper <a href = "https://onlinelibrary.wiley.com/doi/full/10.1111/j.1540-6261.1952.tb01525.x">Portfolio Selection</a> , which was published in the Journal of Finance in 1952. He was later awarded a Nobel Prize for his work on modern portfolio theory.

# Markowitz Model







- We model our assets by their expected return, $E[R]$ and their risk, which is expressed as their standard deviation, $\sigma$

- Our investment decisions are expressed by investing 100% of our wealth in assets( here, stocks), where each particular investment represents a proportion of our total wealth. 


- We will now implement Markowitz Model. This model assists in the selection of the most efficient portfolios by analyzing various possible portfolios of the selected stocks.

- We invest $w_i$ in $stock_i$ for every i, such that $\Sigma^{n}_{i=1} w_i = 1$

- The expected return of the portfolio constructed would be $E[R_p] = \Sigma^{n}_{i=1} w_i E[R_i]$; and the risk associated with the portfolio would be $\sigma^2(R_p) = \Sigma^{n}_{i=1} w_i^2 \sigma^2(R_i) + \Sigma^{}_{i=1}\Sigma^{}_{j {\neq} i} w_i w_j \sigma(R_i) \sigma(R_j) \rho_{ij}$, where $\rho_{ij}$ is the correlation between logarithmic returns of $i$th and $j$th stocks

```python
# A function for generating a numpy array containing random weights that add upto 1
def RandWeights(size):
    weight = np.random.dirichlet(np.ones(size))
    return weight
```

```python

risk_free_rate = 0.0619 # 6.19% on July 9

def portfolio_stats(weight):

    # Convert to array in case list was passed instead.
    weight = np.array(weight)
    port_return = np.sum(log_returns.mean() * weight) * 250
    port_vol = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov() * 250, weight)))
    sharpe = (port_return - risk_free_rate)/port_vol

    return {'return': port_return, 'volatility': port_vol, 'sharpe': sharpe}
```

```python
length = len(log_returns.columns)
weight = np.random.random(length)
weight = weight/np.sum(weight)
weight
```

```python
def Monte_Carlo(iterations):
    portfolio_returns = []
    portfolio_volatilities = []
    for x in range (iterations):
        weight = RandWeights(length)
        pf_stats = portfolio_stats(weight)
        portfolio_returns.append(pf_stats['return'])
        portfolio_volatilities.append(pf_stats['volatility'])
        
    portfolio_returns = np.array(portfolio_returns)
    portfolio_volatilities = np.array(portfolio_volatilities)
    return portfolio_returns, portfolio_volatilities

    
```

```python
portfolio_returns, portfolio_volatilities = Monte_Carlo(100000)
sharpe = portfolio_returns / portfolio_volatilities
max_sr_ret = portfolio_returns[sharpe.argmax()]
max_sr_vol = portfolio_volatilities[sharpe.argmax()]
plt.figure(figsize=(18,10))
plt.scatter(portfolio_volatilities, portfolio_returns, c=sharpe, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_sr_vol, max_sr_ret,c='red', s=50) # red dot
plt.show()
```

```python

def minimize_volatility(weights):  
    return portfolio_stats(weights)['volatility']

```

Using Optimization to find portfolio with max sharpe ratio

```python

def OptimizationWithSharpeRatio():

    def FindNegSharpe(weight):
        return (-1)*portfolio_stats(weight)['sharpe']

    res = minimize(
          FindNegSharpe,
          RandWeights(length),
          constraints=[
            {'type': 'eq', 'fun': lambda w: np.sum(w) - 1.},
          ],
          bounds=[(0., 1.) for i in range(length)]
        )
    
    return res
```

```python
OptimizationWithSharpeRatio()
```

Using Optimisation to find portfolio that has minimum risk for a given expected return

```python
def OptimizationForAGivenReturn(target_return):
    res = minimize(
      minimize_volatility,
      RandWeights(length),
      method = 'SLSQP',
      constraints=[{'type':'eq','fun': lambda x: portfolio_stats(x)['return']-target_return},
                   {'type':'eq','fun': lambda x: np.sum(x)-1}],
      bounds=[(0., 1.) for i in range(length)]
    )
    
    return res
```

```python
OptimizationForAGivenReturn(0.5)
```

Optimizing with risk tolerance

```python
def OptimisingWithTolerance(tolerance):
    
    def fun(weight):
        pf_stats = portfolio_stats(weight)
        _return = pf_stats['return']
        _risk = pf_stats['volatility']
        
        return _risk - tolerance * _return
        
    
    res = minimize(
      fun,
      RandWeights(length),
      method = 'SLSQP',
      constraints=[
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1.},
      ],
      bounds=[(0., 1.) for i in range(length)]
    )

    return res
```

```python
OptimisingWithTolerance(0.1)
```

```python
OptimisingWithTolerance(0.5)
```

Plotting the efficient Frontier

```python
target_returns = np.linspace(portfolio_returns.min(), portfolio_returns.max(),70)

# Initialize optimization parameters
minimal_volatilities = []
for target_return in target_returns:
    optimal = OptimizationForAGivenReturn(target_return)
    minimal_volatilities.append(optimal['fun'])

minimal_volatilities = np.array(minimal_volatilities)
print(minimal_volatilities)
```

```python
plt.figure(figsize=(18,10))

plt.scatter(portfolio_volatilities, portfolio_returns,
            c = ( portfolio_returns / portfolio_volatilities),
            marker = 'o')

plt.scatter(minimal_volatilities,
            target_returns,
            c = (target_returns / minimal_volatilities),
            marker = 'x')

Optimal_weights_For_Highest_Sharpe_Ratio = OptimizationWithSharpeRatio().x

plt.plot(portfolio_stats(Optimal_weights_For_Highest_Sharpe_Ratio)['volatility'],
         portfolio_stats(Optimal_weights_For_Highest_Sharpe_Ratio)['return'],
         'r*',
         markersize = 25.0)

Optimal_weights_For_Lowest_Risk = OptimisingWithTolerance(0).x

plt.plot(portfolio_stats(Optimal_weights_For_Lowest_Risk)['volatility'],
         portfolio_stats(Optimal_weights_For_Lowest_Risk)['return'],
         'y*',
         markersize = 25.0)

Optimal_weights_for_twenty_percent_tolerance = OptimisingWithTolerance(1).x

plt.plot(portfolio_stats(Optimal_weights_for_twenty_percent_tolerance)['volatility'],
         portfolio_stats(Optimal_weights_for_twenty_percent_tolerance)['return'],
         'g*',
         markersize = 25.0)

Optimal_weights_for_thirty_five_percent_returns = OptimizationForAGivenReturn(0.35).x

plt.plot(portfolio_stats(Optimal_weights_for_thirty_five_percent_returns)['volatility'],
         portfolio_stats(Optimal_weights_for_thirty_five_percent_returns)['return'],
         'b*',
         markersize = 25.0)


plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.colorbar(label='Sharpe ratio')
```
